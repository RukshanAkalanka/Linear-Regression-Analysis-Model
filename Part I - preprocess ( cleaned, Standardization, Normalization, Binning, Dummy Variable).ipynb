```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('Auto_vehicle.csv', header=None)
df.head()
```

```python
df.shape
df.tail(10)
```

```python
#add column names to dataframe
column_names= ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns
df.columns= column_names
df.head(20)
```

```python
#data wrangling / data cleaning
df.replace("?", np.nan, inplace=True)
df.head()
```

```python
#evaluate for mising data idnull() and notnull
df.isnull().sum()
```

```python
#check data type of each column in data frame
df.dtypes
```

```python
#change the colun data type and replace null (NaN) values with appropriate values
avg_norm_loss= df["normalized-losses"].astype("float").mean()
avg_bore= df["bore"].astype("float").mean()
avg_stroke= df["stroke"].astype("float").mean()
avg_peak_rpm= df["peak-rpm"].astype("float").mean()
avg_horsepower= df["horsepower"].astype("float").mean()

print("average normal loss : ",avg_norm_loss)
print("average bore : ",avg_bore)
print("average stroke : ",avg_stroke)
print("average peak-rpm : ",avg_peak_rpm)
print("average horsepower : ",avg_horsepower)
```

```python
#replace NaN values by mean
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
df["bore"].replace(np.nan, avg_bore, inplace=True)
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
df["peak-rpm"].replace(np.nan, avg_peak_rpm, inplace=True)
df["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
df
```

```python
#replace num-of-doors nan values using frequency
df["num-of-doors"].value_counts()
df["num-of-doors"].replace(np.nan, "four", inplace=True)
df.head()
```

```python
#cleaning NaN values in price column
# that column is the impotent column beacuse that is we will use to prediction value in our model. so we need to drop every row when 
# price column has NaN value

df.dropna(subset=["price"], axis=0, inplace=True)

#reset left hand side indises, beacuse we droped two rows
df.reset_index(drop=True, inplace=True)
df.head()

df.isnull().sum()
```

```python
df.dtypes
#need to convert data types to proper format
```

```python
df[["bore","stroke","peak-rpm","price"]]= df[["bore","stroke","peak-rpm","price"]].astype("float")
df[["normalized-losses"]]= df[["normalized-losses"]].astype("int")
df.dtypes
df.to_csv("cleaned_Auto_vehicle.csv", index=False)
```

```python
#  (Standardization, Normalization, Binning, Dummy Variable)

#load clean dataset
df= pd.read_csv("cleaned_Auto_vehicle.csv")
df.head()
```

```python
# data Standardization : Standardization is the process of transforming data into a common format which allows the researcher to make the meaningful comparison.
#in my data frame city-mpg and highway-mpg present in miles per galum. but we want liters per 100 km we need to unit convertion

#1 Gallon = 3.785 Liter, 1 Mile = 1.609 Kilometer, L/1 km = 3.785/1,609 = 2.35 mpg, L/100km = 2.35*100 = 235

#convert mpg to L/100km by mathematical operation(235 devided by mpg)
df["highway-L/100km"]= (235/df["highway-mpg"])
df["city-L/100km"]= (235/df["city-mpg"])
df.head()
df.dtypes[["highway-L/100km","city-L/100km"]]
```

```python
#Normalized data: Some data column spred big range. so we nedd to normalize that column into 0 to 1 range
#replace original value by (original value)/(maximum value)
df[["length","width","height"]].head()
```

```python
df["length"]= df["length"]/df["length"].max()
df["width"]= df["width"]/df["width"].max()
df["height"]= df["height"]/df["height"].max()
df[["length","width","height"]].head()
```

```python
#Binning technique
#you know vehicle price range depend on housepower low, medium and high. so no need to each and every vehical exact house power 
#beacuse of that we need to conver housepower into 3 different category we called it binnning. in dataframe max is 262 and min : 48

#we will use the pandas method 'cut' to segment housepower column into 3 bins

# 3 bins of equal size bandwidth so we use numpy's linspace(start value, end value, numbers generated) function

#we are building 3 bins of equal length, there should be 4 dividers. so numbers_generated=4

bins= np.linspace(df["horsepower"].min(), df["horsepower"].max(), 4)
bins
```

```python
#set the group names
group_names= ["low","medium","high"]

#apply pandas cut method determine each value of df["horsepower"] belogs to
df["horsepower-binned"]= pd.cut(df["horsepower"], bins, labels=group_names, include_lowest=True)
df[["horsepower","horsepower-binned"]].head(20)
```

```python
#see value count for each type
df["horsepower-binned"].value_counts()
```

```python
import matplotlib as plt
from matplotlib import pyplot

group_names=["low","medium","high"]
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

#set x and y labels and plot title
plt.pyplot.xlabel("hoursepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("hoursepower bins")
```

```python
#indicator varibles/ dummy variables 

#our target is liner regression model for that we need to that categorical varibles. so some cotegorical varibles ("labled : ex medium,low, high, fuel-type : gas , diesel) we need to conver numbers
#that are called ''dummy' because the numbers themselves don't have inherent meaning

# exaple : in this dataset fuel-type has two values that is "gas" and "diesel" only. but in regression model they dont understand the word for that 
# we need to use numbers for fuel-type. so need to convert fuel-type into indicator varibles

#for create dummy varibles we use pandas "get_dummies" method to assign numerical values of different category of fuel-types

df.columns
```

```python
df.head(20)
```

```python
dummy_variable_1= pd.get_dummies(df["fuel-type"])
dummy_variable_1.head(20)
```

```python
#see fuel type value_counts()
df["fuel-type"].value_counts()
```

```python
#rename dummy_varible_1 data frame coumn names
dummy_variable_1.rename(columns={"gas": "fuel-type-gas", "diesel": "fuel-type-diesel"}, inplace=True)
dummy_variable_1.head(20)
```

```python
#you know dummy_variables_1 dataframe is a temporary separate dataframe so we need to merge that into our main df dataframe

df= pd.concat([df, dummy_variable_1], axis=1) #axis=1 mean that dataframe add column wise
df.drop("fuel-type", axis=1, inplace=True) #drop original fuel type column
df.head(20)
```
